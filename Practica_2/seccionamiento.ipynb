## Hay que hacer dos cosas, la primera es obtener un horizonte de 500 muestras y otro de 20

### Seg√∫n lo platicado en clase, entonces vamos a seguir las siguientes proporciones para la serie:

* Parte 1 
    * Horizonte: 500
    * Ventana  : 1000

* Parte 2
    * Horizonte: 20
    * Ventana  : 200 (para aprovechar que se pueden usar mas datos)


```python
ruta = 'data/raw_data/nasdaq.csv'
df = pd.read_csv(ruta , names = ['Valores'])
df.head()
```

### Seccionamiento Parte 1

```python
ventana_1 = df[:1000]
horizonte_1 = df[1000:]
```

```python
ventana_1.tail()
```

```python
horizonte_1.head()
```

```python
ventana_1.to_csv('data/clean_data/v1.csv', index = True)
horizonte_1.to_csv('data/clean_data/h1.csv', index = True)
```

### Seccoinamiento Parte 2

```python
ventana_2 = df[-200:-20]
horizonte_2 = df[-20:]
```

```python
horizonte_2.head()
```
